```python
# FInd the min, max, mean, median, std, skew and kurt of the relevent numerical columns.

def uni_num_analysis(num_data):
    for col_name in num_data:
        print('*'*10,col_name,'*'*10)
        stats = num_data[col_name].agg(['min','max','mean','median','std','skew','kurt'])
        stats_rounded = stats.round(2)
        print(stats_rounded)
        print()
```

```python
uni_num_analysis(df[['Salary']])
```

```python
#  Find the unique, nunique, count of the relevent categorical columns.

def uni_cat_analysis(cat_data):
    for col_name in cat_data:
        print('*'*10,col_name,'*'*10)
        print(cat_data[col_name].agg(['unique', 'nunique','count']))
        print('Value Counts: \n', cat_data[col_name].value_counts())
        print()
```

```python
# Find the outliers in each numerical columns:

def find_outliers(df):
    outliers = {}
    for col in df.select_dtypes(include=['number']).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 -Q1

        lower_fence = Q1 - 1.5 * IQR
        upper_fence = Q3 + 1.5 * IQR

        outliers[col] = df[(df[col] < lower_fence) | (df[col] > upper_fence)][col]

        print(f"Outliers in {col}:")
        print(outliers[col].values)
        print('-' * 30)
    
    return outliers
```

```python
find_outliers(df)
```

```python
# Using zscore Method
from scipy.stats import zscore

ol ={}
for col in df.select_dtypes(include=['number']).columns:
    z_score = zscore(df[col])
    ol[col] = df[col][abs(z_score) > 3]
    print(f"Outliers in {col}")
    print(ol[col].values)
    print('-'*20)
```

```python
# Treating the ourliers. As salary is the target variable, we can treat the outliers in the salary column.

q1 = df['Salary'].quantile(0.25)
q3 = df['Salary'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 + iqr

df1 = df[(df['Salary'] > lower_bound) & (df['Salary'] < upper_bound)]
```

We have drop the outliers from the dataset in terms of the 'Salary' variable.

```python
def scatter_plot(df, col1, col2):
    # plt.figure(figsize=(6,4))
    sns.scatterplot(data=df, x=col1,y=col2)
    plt.title(f"Scatter Plot: {col1} Vs {col2}")
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.tight_layout()
    # plt.show()
```

```python
plt.figure(figsize=(16,12))
n = 1
for i in ['10percentage', '12percentage', 'collegeGPA','English', 'Logical', 'Quant']:
    plt.subplot(3,3,n)
    scatter_plot(df1,'Salary',i)
    n +=1
plt.show()
```

```python
def hexbin_plot(df, col1, col2):
    # plt.figure(figsize=(4,3))
    plt.hexbin(df[col1], df[col2], gridsize=10, cmap='Blues')
    plt.colorbar(label='Counts')
    plt.title(f"Hexbin Plot: {col1} Vs {col2}")
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.tight_layout()
    # plt.show()
```

```python
plt.figure(figsize=(16,12))
n = 1
for i in ['10percentage', '12percentage', 'collegeGPA','GraduationYear', 'English', 'Logical', 'Quant']:
    plt.subplot(3,3,n)
    hexbin_plot(df1,'Salary',i)
    n+=1
plt.show()
```

Identify the patterns between categorical and numerical columns using swarmplot, boxplot, barplot, etc.

```python
# Box Plot

def box_plot(df, x_col, y_col):
    sns.boxplot(data=df,x=x_col,y=y_col)
    plt.title(f"Box Plot: {y_col} by {x_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.tight_layout()
```

```python
plt.figure(figsize=(16,8))
n = 1
for i in ['Gender','Degree']:
    plt.subplot(1,2,n)
    box_plot(df1,i, 'Salary')
    n += 1
plt.show()
```

The box plot 'Salary Vs Gender' shows that the median salary for both the male and female is same. Male shows higher range of salary values.

The Box plot comparing Salary and Degree shows that M.Sc (Tech) graduates have the highest median salary. M.Tech / M.E. have the wider salary range.

```python
# Swarm Plot

def box_plot(df, x_col, y_col):
    sns.swarmplot(data=df,x=x_col,y=y_col,size=0.5,edgecolor='auto')
    plt.title(f"Swarm Plot: {y_col} by {x_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.tight_layout()
```

```python
plt.figure(figsize=(16,8))
n = 1
for i in ['Gender','Degree']:
    plt.subplot(1,2,n)
    box_plot(df1,i, 'Salary')
    n += 1
plt.show()
```

In the first swarm plot, it shows the salary distribution of males is more spread as compared to the female. The other swarm plot show the salary distribution different degree holders. We can clearly see that, the salary distribution of B.Tech/B.E. graduate is more spread out with several high earning indivisuals while M.Sc.(Tech) graduates have more tighlty clustered distribution, indication less variation in salary.

```python
# Bar Plot

def bar_plot(df, x_col, y_col):
    sns.barplot(data=df,x=x_col,y=y_col)
    plt.title(f"Bar Plot: {y_col} by {x_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.tight_layout()

plt.figure(figsize=(16,8))
n = 1
for i in ['Gender','Degree']:
    plt.subplot(1,2,n)
    bar_plot(df1,i, 'Salary')
    n += 1
plt.show()
```

The above bar plot 'Salary Vs Gender' reveals that both the male and female earn the same salary.

The 'Salary Vs Degree' bar plot shows that the M.Sc.(Tech) graduates, on average, earn the highest salary, while MCA degree holders have the lowest.

```python
# Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

# Creating a table for gender and specialization
gender_table = pd.crosstab(df1['Specialization'],df1['Gender'])

# Perform the chi-square test to determine if there is any dependancy between gender or specialization.

from scipy.stats import chi2_contingency

# chi2 Test

chi2_stat, p_value, dof, expected = chi2_contingency(gender_table)

print(f"Chi-Square Statistic : {chi2_stat}")
print(f"P-value : {p_value}")
print(f"Degree of Freedom : {dof}")
```

The Chi_squared test gave the result of P_value of 1.0235850782721279e-06, which is much smaller than the significance level of 0.05.
Therefor we reject the null hypothesis and conclude that there is a significance relationship between Gender and Specialization. This means, specializations choosen by the candidates is influenced by the gender.
